```python
import json
import numpy as np
import pandas as pd

dict_countries = []

with open('owid-co2-data.json', 'r') as file:
    data = json.load(file)
    for country, values in data.items():
        for year in values['data']:
            dict_countries.append([country, values.get('iso_code',np.nan), year['year'], year.get('co2', np.nan),  year.get('population', np.nan)])

dict_countries
```

```python
df = pd.DataFrame(dict_countries, columns=['country','year','iso_code','C02','population'])
df
```
