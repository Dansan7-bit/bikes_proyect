```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df_ev = df.dropna(subset=['ev_shares'])
df_ev
```

```python
df_ev['country'].unique()
```

```python
countries_continents = {
    'Australia': 'Oceania',
    'Austria': 'Europe',
    'Belgium': 'Europe',
    'Brazil': 'South America',
    'Canada': 'North America',
    'Chile': 'South America',
    'China': 'Asia',
    'Denmark': 'Europe',
    'Finland': 'Europe',
    'France': 'Europe',
    'Germany': 'Europe',
    'Greece': 'Europe',
    'Iceland': 'Europe',
    'India': 'Asia',
    'Israel': 'Asia',
    'Italy': 'Europe',
    'Japan': 'Asia',
    'Mexico': 'North America',
    'Netherlands': 'Europe',
    'New Zealand': 'Oceania',
    'Norway': 'Europe',
    'Poland': 'Europe',
    'Portugal': 'Europe',
    'South Africa': 'Africa',
    'South Korea': 'Asia',
    'Spain': 'Europe',
    'Sweden': 'Europe',
    'Switzerland': 'Europe',
    'Turkey': 'Europe',
    'United Kingdom': 'Europe',
    'United States': 'North America'
}
df_ev['continent'] = df_ev['country'].map(countries_continents)
df_ev
```

```python
df_ev_group = df_ev.groupby(['continent','year'])[['C02','GDP','ev_shares','population']].mean().reset_index()
df_ev_group
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo de seaborn
sns.set(style="whitegrid")

# Listado de métricas a graficar
metrics = ['C02', 'population', 'GDP', 'ev_shares']

# Tamaño de la figura
plt.figure(figsize=(15, 10))

# Loop para crear subgráficas
for i, metric in enumerate(metrics, 1):
    plt.subplot(2, 2, i)
    sns.lineplot(data=df_ev_group, x='year', y=metric, hue='continent')
    plt.title(f'Trend of {metric} by Continent')
    plt.xlabel('Year')
    plt.ylabel(metric)
    plt.legend(title='Continent', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar el layout
plt.tight_layout()
plt.show()
```

```python
df_ev_group.reset_index(inplace=True, drop=True)
df_ev_group
```

```python
x_num = df_ev_group[['C02','population','GDP','ev_shares']]
x_num
```

```python

from sklearn.preprocessing import MinMaxScaler
x_num = df_ev_group[['C02','population','GDP','ev_shares']]
tranformer = MinMaxScaler().fit(x_num)
x_normalize = tranformer.transform(x_num)
# x_normalize --> Te devuelve todas las columnas normalizadas, pero en un array
print(x_normalize.shape)
df_normalize = pd.DataFrame(x_normalize, columns=x_num.columns)
df_normalize.reset_index(inplace=True)
df_normalize
```

```python
df_ev_group.reset_index(inplace=True)
```

```python
df1 = pd.merge(df_ev_group, df_normalize, on='index')
```

```python
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo de seaborn
sns.set(style="whitegrid")

# Listado de métricas a graficar
metrics = ['C02_y', 'population_y', 'GDP_y', 'ev_shares_y']

# Tamaño de la figura
plt.figure(figsize=(15, 10))

# Loop para crear subgráficas
for i, metric in enumerate(metrics, 1):
    plt.subplot(2, 2, i)
    sns.lineplot(data=df1, x='year', y=metric, hue='continent')
    plt.title(f'Trend of {metric} by Continent')
    plt.xlabel('Year')
    plt.ylabel(metric)
    plt.legend(title='Continent', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar el layout
plt.tight_layout()
plt.show()
```

```python
# HEAT MAP CORRELATION
df1 = df_ev_group[df_ev_group['continent'] == 'Asia']
df1.drop(columns=['index'], inplace=True)
numeric_columns = df1.select_dtypes(include=[float, int]).columns
grouped_df = df1.groupby(['continent', 'year'],as_index=False)[numeric_columns].mean()
grouped_df.drop(columns=["year"], inplace=True)
num = grouped_df.select_dtypes(include=[float, int])
correlations_matrix = num.corr()
correlations_matrix = num.corr()
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```
