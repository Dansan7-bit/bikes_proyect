```python
import requests
import pandas as pd

# URL base de la API para obtener indicadores
base_url = 'https://api.worldbank.org/v2/indicators'
params = {
    'format': 'json',
    'per_page': 10000  # Número de resultados por página, máximo permitido por la API
}

# Lista para almacenar todos los indicadores
all_indicators = []

page = 1
while True:
    params['page'] = page
    response = requests.get(base_url, params=params)
    data = response.json()

    # Añadir los indicadores a la lista
    indicators = data[1]
    all_indicators.extend(indicators)
    
    # Verificar si hay más páginas
    if len(indicators) == 0:
        break
    
    page += 1

# Convertir la lista de indicadores en un DataFrame de Pandas
df_indicators = pd.DataFrame(all_indicators)

# Mostrar los primeros registros del DataFrame
print(df_indicators.head())

# Guardar los indicadores en un archivo CSV
df_indicators.to_csv('worldbank_indicators.csv', index=False)
```
